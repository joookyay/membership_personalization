```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
#데이터 불러오기

df_1 = pd.read_csv(r"LPOINT_BIG_COMP_01_DEMO.csv")
df_2 = pd.read_csv(r"LPOINT_BIG_COMP_02_PDDE.csv")
df_3 = pd.read_csv(r"LPOINT_BIG_COMP_03_COP_U.csv")
df_4 = pd.read_csv(r"LPOINT_BIG_COMP_04_PD_CLAC.csv")
df_5 = pd.read_csv(r"LPOINT_BIG_COMP_05_BR.csv")
df_6 = pd.read_csv(r"LPOINT_BIG_COMP_06_LPAY.csv")
```

```python
#데이터 합치기 

df_total = pd.concat([df_1, df_2], ignore_index = True)
df_total = pd.concat([df_total, df_3], ignore_index = True).drop_duplicates()
df_total = pd.merge(df_total, df_4, left_on='pd_c', right_on='pd_c', how='left')
```

```python
#na값 제거

df_total_new = df_total.dropna(subset=['pd_nm'])


#df_total_new.to_csv("df_total_new.csv", index = False)
```

```python
#연관분석을 위한 데이터프레임 만들기

b = pd.DataFrame(df_total_new.groupby('cust')['pd_nm'].apply(list)).reset_index()['pd_nm']


te = TransactionEncoder()
te_ary = te.fit(b).transform(b)
df2 = pd.DataFrame(te_ary, columns=te.columns_)
df2.head()
```

```python
#최소 지지도 0.15 연관분석 결과

from mlxtend.frequent_patterns import apriori
frequent_item=apriori(df2,min_support=0.15,use_colnames=True)
frequent_item  
```

```python
#최소 신뢰도 0.8 연관분석 결과

from mlxtend import frequent_patterns
from mlxtend.frequent_patterns import association_rules
store_association=association_rules(frequent_item,metric="confidence",min_threshold=0.8)
store_association
```

```python
store_association_new=store_association[['antecedents', 'consequents', 'antecedent support',
       'consequent support', 'support', 'confidence', 'lift']]
```

```python
#연관분석 최종 결과 출력

store_association_new
```
